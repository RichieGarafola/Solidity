**Solidity and Remix**
Solidity code is consdiered particularly human readable with characteristics that resemble those of both Python and JavaScript
When a soliditiy smart contract runs, the program first gets compiled.
This means that its onverted from human-readable syntax into a computer-friendly version of the code, called bytecode
once compiled, the code runs in the EVM (Ethereum Virtual Machine)

---

The EVM is a software platform that is embedded in each node of the Ethereum blockchain network. The EVM runs the bytecode that results from running a solidity smart contract. 

---

**A key benefit of using the EVM**
The EVM offers an isolated enviornment in which Solidity code an run - without accessing the computer network, files, or other resources of the host computer.
This means that by using the same instructions, each decentralized node on the Ethereum blockchain can validate the smart contract. The EVM is thus essential to the consensus mechanism of the Etherem blockchain.

---

In general, a consensus mechanism allows the nodes of the Ethereum blockchin to work together to both stay secure and reach an agreement on the current state of the blockchain.
